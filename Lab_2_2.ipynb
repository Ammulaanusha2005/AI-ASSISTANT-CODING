# **Task 1: Cleaning sensor** **Data**


### #Use Gemini in Google Colab to generate a Python function that filters out all negative numbers from a list of sensor readings.

```python
def remove_negative_values(sensor_data):
    cleaned_data = []
    for value in sensor_data:
        if value >= 0:
            cleaned_data.append(value)
    return cleaned_data
# Sample IoT sensor data
sensor_data = [12, -5, 8, -3, 0, 15, -10]
print("Before cleaning:", sensor_data)
print("After cleaning:", remove_negative_values(sensor_data))
```

# **Task 2:String Character Analysis**

### #Use Gemini to generate a Python function that counts vowels, consonants,and digits in a string.

```python
def count_characters(text):
    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0
    digit_count = 0
    for ch in text:
        if ch.isdigit():
            digit_count += 1
        elif ch.isalpha():
            if ch in vowels:
                vowel_count += 1
            else:
                consonant_count += 1
    return vowel_count, consonant_count, digit_count
```

### sample input & output:

```python
text = "Hello123World"
vowels, consonants, digits = count_characters(text)
print("String:", text)
print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
```

# **Task 3:Palindrome Check-Tool Comparison**

### #Generate a Python function to check whether a given string is a palindrome.



Code generated by Gemini:

```python
def is_palindrome(text):
    text = text.lower()
    reversed_text = text[::-1]
    return text == reversed_text
```

Code generated by Copilot:

```python
def is_palindrome(text):
    cleaned = ''.join(char.lower() for char in text if char.isalnum())
    left = 0
    right = len(cleaned) - 1
    while left < right:
        if cleaned[left] != cleaned[right]:
            return False
        left += 1
        right -= 1
    return True
```
