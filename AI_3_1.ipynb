1:Zero-Shot Prompting (Palindrome Number Program)

### #Write a Python function to check whether a given number is a palindrome

AI-Genarated code:

```python
n = int(input("Enter a number: "))
s = str(n)

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

input:183->Output:False

Observation: It works for positive numbers and it doesnot handle negative numbers

### 2: One-Shot Prompting (Factorial Calculation)

#Write a Python function to calculate factorial of a number.Example: Input: -4 → Output: Invalid

AI-Generated Code:

```python
def factorial(n):
    if n < 0:
        return "Invalid input"
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

n = int(input("Enter a number: "))
print(factorial(n))
```

Testing:

input:3->output:6

input:-4->output:Invalid

Observation: Better clarity than zero-shot method

3:Few-Shot Prompting (Armstrong Number Check)
# Write a Python function to check whether a number is an Armstrong number.

Examples: 371→ Armstrong, 470 → Armstrong, 123 → Not Armstrong

AI-Generated Code:

```python
n = int(input("Enter a number: "))
s = str(n)
total = 0

for i in s:
    total += int(i) ** len(s)

if total == n:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

Testing:

input: 370->Armstrong Number

input: 123->Not an Armstrong Number

Observation: Few-shot examples improved accuracy and logic of a code

5: Zero-Shot Prompting (Perfect Number Check)
# Write a Python function to check whether a given number is a perfect number
AI-Generated Code:

```python
n = int(input("Enter a number: "))

if n <= 1:
    print("Neither Prime nor Composite")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Composite")
            break
    else:
        print("Prime")
n = int(input("Enter a number: "))
total = 0

for i in range(1, n):
    if n % i == 0:
        total += i

if total == n:
    print("Perfect Number")
else:
    print("Not a Perfect Number")
```

Testing:

18=Not a Perfect Number
97=Not a Perfect Number

Observation: Logic correct but inefficient for large numbers

# Write a Python program to determine whether a number is even or odd with input validation.
